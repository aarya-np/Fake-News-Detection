```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
```

```python
import csv
import json

with open('F.csv','r',encoding='utf-8') as load_file:
    data=csv.DictReader(load_file)
    new_data=[]
    for row in data:
        new_data.append(row)
with open('fake.json','w') as file:
    json.dump(new_data, file, indent=2)
```

```python
import json
import csv
with open('NF.csv','r',encoding='utf-8') as load_file:
    data=csv.DictReader(load_file)
    new_data=[]
    for row in data:
        new_data.append(row)
with open('notfake.json','w')as file:
    json.dump(new_data, file, indent=2)
```

```python
with open('fake.json','r') as file:
    dataset1=json.load(file)
fake_sentences=[]
fake_labels=[]
for item in dataset1:
    fake_sentences.append(item['title'])
for i in range(len(fake_sentences)):
    fake_labels.append(1)
print(len(fake_sentences),len(fake_labels))
```

```python
with open('notfake.json','r') as file:
    dataset1=json.load(file)
true_sentences=[]
true_labels=[]
for item in dataset1:
    true_sentences.append(item['title'])
for i in range(len(true_sentences)):
    true_labels.append(0)
print(len(true_sentences),len(true_labels))
```

```python
from sklearn.model_selection import train_test_split

# Assuming fake_sentences, true_sentences, fake_labels, and true_labels are defined

# Splitting fake data
train_fake, test_fake, trainfake_labels, testfake_labels = train_test_split(fake_sentences, fake_labels, test_size=0.2, random_state=42)

# Splitting true data
train_true, test_true, traintrue_labels, testtrue_labels = train_test_split(true_sentences, true_labels, test_size=0.2, random_state=42)

# Displaying lengths
print(len(train_fake), len(test_fake), len(trainfake_labels), len(testfake_labels))
print(len(train_true), len(test_true), len(traintrue_labels), len(testtrue_labels))
```

```python
train_sentences=train_fake+train_true
test_sentences=test_fake+test_true

train_labels=trainfake_labels+traintrue_labels
test_labels=testfake_labels+testtrue_labels

print(len(train_sentences), len(train_labels), len(test_sentences), len(test_labels))
```

```python
train_labels=np.array(train_labels)
test_labels=np.array(test_labels)
vocab_size=30000
embedded_dim=16
max_length=25
trunc_type='post'
padding_type='post'
oov_token='<OOV>'
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
tokenizer=Tokenizer(num_words=vocab_size,oov_token=oov_token)
tokenizer.fit_on_texts(train_sentences)
word_index=tokenizer.word_index
print(word_index)
```

```python
train_sequence=tokenizer.texts_to_sequences(train_sentences)
train_padded=pad_sequences(train_sequence, maxlen=max_length,padding=padding_type,truncating=trunc_type)
```

```python
test_sequence=tokenizer.texts_to_sequences(test_sentences)
test_padded=pad_sequences(test_sequence, maxlen=max_length,padding=padding_type,truncating=trunc_type)
```

```python
model=tf.keras.models.Sequential([tf.keras.layers.Embedding(vocab_size, embedded_dim, input_length=max_length),
                                 tf.keras.layers.Flatten(),
                                 tf.keras.layers.Dense(24, activation='relu'),
                                 tf.keras.layers.Dense(1,activation='sigmoid')])
model.build()
```

```python
model.summary()
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
             optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
             metrics=['accuracy'])
history=model.fit(train_padded, train_labels, epochs=15, validation_data=(test_padded,test_labels))
```

```python
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']

epochs=range(len(acc))
plt.plot(epochs, acc)
plt.plot(epochs, val_acc)
plt.figure()

plt.plot(epochs,loss)
plt.plot(epochs,val_loss)
plt.figure()
```

```python
sentence=['Minsk cultural hub becomes haven from authorities','How to Blow $700 Million: Al Jazeera America Finally Calls it Quits','Indonesia to buy $1.14 billion worth of Russian jets','JUSTICE? Yahoo Settles E-mail Privacy Class-action: $4M for Lawyers, $0 for Users']
test_sen=tokenizer.texts_to_sequences(sentence)
test_padded=pad_sequences(test_sen, maxlen=max_length,padding=padding_type,truncating=trunc_type)
```

```python
model.predict(test_padded)
```
